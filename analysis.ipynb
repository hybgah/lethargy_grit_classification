```r
# 역코딩 함수 정의
# 입력값: 데이터 테이블, 역코딩할 열의 이름이 담긴 벡터
# 출력값: 역코딩이 완료된 데이터 테이블
reverse_code <- function(data, cols_to_reverse) {
  # 데이터가 data.table인지 확인
  if (!"data.table" %in% class(data)) {
    stop("Input data must be a data.table")
  }

  # 입력된 역코딩 열이 데이터에 존재하는지 확인
  missing_cols <- setdiff(cols_to_reverse, names(data))
  if (length(missing_cols) > 0) {
    stop(paste("The following columns are missing in the data.table:", paste(missing_cols, collapse = ", ")))
  }

  # 역코딩 수행: 5 - 값으로 계산
  data[, (cols_to_reverse) := lapply(.SD, function(x) 5 - x), .SDcols = cols_to_reverse]

  # 역코딩 완료된 데이터 반환
  return(data)
}
```

```r
# NA 값 확인 함수 정의
check_na_in_tables <- function(data_tables, verbose = TRUE) {
  # 리스트 형식 확인
  if (!is.list(data_tables)) {
    stop("Input must be a list of data.table objects")
  }

  # 테이블 이름 가져오기
  table_names <- names(data_tables)

  # 테이블 이름이 없는 경우 자동 이름 지정
  if (is.null(table_names) || any(table_names == "")) {
    table_names <- paste0("Table_", seq_along(data_tables))
  }

  # 결과 저장 리스트
  results <- list()

  # 각 테이블에서 NA 값 확인
  for (i in seq_along(data_tables)) {
    dt <- data_tables[[i]]
    table_name <- table_names[i]

    # 데이터가 data.table인지 확인
    if (!"data.table" %in% class(dt)) {
      stop(paste(table_name, "is not a data.table"))
    }

    # NA 값 확인
    na_count <- sum(is.na(dt))
    na_exists <- na_count > 0

    # 결과 저장
    results[[table_name]] <- list(na_exists = na_exists, na_count = na_count)

    # 선택적 출력
    if (verbose) {
      if (na_exists) {
        cat(sprintf("NA values found in %s: %d NA(s)\n", table_name, na_count))
      } else {
        cat(sprintf("No NA values in %s\n", table_name))
      }
    }
  }

  # 결과 반환
  return(results)
}
```

```r
# 함수 호출
result <- check_na_in_tables(data_tables)

# 결과 확인
print(result)
```

```r
add_combined_columns <- function(dataset, column_name_vectors) {
  # column_name_vectors는 리스트 형식이어야 함
  if (!is.list(column_name_vectors)) {
    stop("The column_name_vectors parameter must be a list of character vectors.")
  }

  for (col_names in column_name_vectors) {
    # 접두어 추출
    prefixes <- unique(sub("\\..*$", "", col_names))


    new_column_name <- prefixes[1]  # 접두어를 새로운 열 이름으로 사용

    # 열 이름이 데이터셋에 모두 있는지 확인
    if (!all(col_names %in% names(dataset))) {
      stop(paste("Some column names in", new_column_name, "do not exist in the dataset."))
    }

    # 새로운 열 추가
    dataset[, (new_column_name) := rowSums(.SD, na.rm = TRUE), .SDcols = col_names]
  }

  return(dataset)
}
com_self_feat = c("Self_Data.YPSY3A01w1","Self_Data.YPSY3A02w1","Self_Data.YPSY3A03w1","Self_Data.YPSY3A04w1",
                "Self_Data.YPSY3A05w1","Self_Data.YPSY3A06w1","Self_Data.YPSY3A07w1","Self_Data.YPSY3A08w1",
                "Self_Data.YPSY3A09w1","Self_Data.YPSY3A10w1")
com_attention_feat = c("Attention_Data.YPSY4A01w1","Attention_Data.YPSY4A02w1","Attention_Data.YPSY4A03w1","Attention_Data.YPSY4A04w1",
                      "Attention_Data.YPSY4A05w1","Attention_Data.YPSY4A06w1","Attention_Data.YPSY4A07w1")
com_attd_feat = c("Parents_Attitude_Data.YFAM2A01w1","Parents_Attitude_Data.YFAM2A02w1","Parents_Attitude_Data.YFAM2A03w1","Parents_Attitude_Data.YFAM2A04w1",
              "Parents_Attitude_Data.YFAM2B01w1","Parents_Attitude_Data.YFAM2B02w1","Parents_Attitude_Data.YFAM2B03w1","Parents_Attitude_Data.YFAM2B04w1", # 역채점
              "Parents_Attitude_Data.YFAM2C01w1","Parents_Attitude_Data.YFAM2C02w1","Parents_Attitude_Data.YFAM2C03w1","Parents_Attitude_Data.YFAM2C04w1",
              "Parents_Attitude_Data.YFAM2D01w1","Parents_Attitude_Data.YFAM2D02w1","Parents_Attitude_Data.YFAM2D03w1","Parents_Attitude_Data.YFAM2D04w1", # 역채점
              "Parents_Attitude_Data.YFAM2E01w1","Parents_Attitude_Data.YFAM2E02w1","Parents_Attitude_Data.YFAM2E03w1","Parents_Attitude_Data.YFAM2E04w1",
              "Parents_Attitude_Data.YFAM2F01w1","Parents_Attitude_Data.YFAM2F02w1","Parents_Attitude_Data.YFAM2F03w1","Parents_Attitude_Data.YFAM2F04w1") # 역채점



result = add_combined_columns(combined_data,list(com_self_feat,com_attention_feat,com_attd_feat))
result %>% head()
```

```r
add_mean_columns <- function(dataset, column_name_vectors) {
  # column_name_vectors는 리스트 형식이어야 함
  if (!is.list(column_name_vectors)) {
    stop("The column_name_vectors parameter must be a list of character vectors.")
  }

  for (col_names in column_name_vectors) {
    # 접두어 추출
    prefixes <- unique(sub("\\..*$", "", col_names))

    if (length(prefixes) != 1) {
      stop("The provided column names in one of the vectors do not share a consistent prefix.")
    }

    new_column_name <- paste0(prefixes[1], "_Mean")  # 접두어에 "_Mean"을 붙여 새로운 열 이름 설정

    # 열 이름이 데이터셋에 모두 있는지 확인
    if (!all(col_names %in% names(dataset))) {
      stop(paste("Some column names in", new_column_name, "do not exist in the dataset."))
    }

    # 새로운 열 추가 (평균 계산)
    dataset[, (new_column_name) := round(rowMeans(.SD, na.rm = TRUE),2), .SDcols = col_names]
  }

  return(dataset)
}

# 자아존중감
mean_self_feat = c("Self_Data.YPSY3A01w1","Self_Data.YPSY3A02w1","Self_Data.YPSY3A03w1","Self_Data.YPSY3A04w1",
                "Self_Data.YPSY3A05w1","Self_Data.YPSY3A06w1","Self_Data.YPSY3A07w1","Self_Data.YPSY3A08w1",
                "Self_Data.YPSY3A09w1","Self_Data.YPSY3A10w1")

# 주의집중
mean_attention_feat = c("Attention_Data.YPSY4A01w1","Attention_Data.YPSY4A02w1","Attention_Data.YPSY4A03w1","Attention_Data.YPSY4A04w1",
                      "Attention_Data.YPSY4A05w1","Attention_Data.YPSY4A06w1","Attention_Data.YPSY4A07w1")

# 부모의 양육태도
mean_attd_feat = c("Parents_Attitude_Data.YFAM2A01w1","Parents_Attitude_Data.YFAM2A02w1","Parents_Attitude_Data.YFAM2A03w1","Parents_Attitude_Data.YFAM2A04w1",
              "Parents_Attitude_Data.YFAM2B01w1","Parents_Attitude_Data.YFAM2B02w1","Parents_Attitude_Data.YFAM2B03w1","Parents_Attitude_Data.YFAM2B04w1", # 역채점
              "Parents_Attitude_Data.YFAM2C01w1","Parents_Attitude_Data.YFAM2C02w1","Parents_Attitude_Data.YFAM2C03w1","Parents_Attitude_Data.YFAM2C04w1",
              "Parents_Attitude_Data.YFAM2D01w1","Parents_Attitude_Data.YFAM2D02w1","Parents_Attitude_Data.YFAM2D03w1","Parents_Attitude_Data.YFAM2D04w1", # 역채점
              "Parents_Attitude_Data.YFAM2E01w1","Parents_Attitude_Data.YFAM2E02w1","Parents_Attitude_Data.YFAM2E03w1","Parents_Attitude_Data.YFAM2E04w1",
              "Parents_Attitude_Data.YFAM2F01w1","Parents_Attitude_Data.YFAM2F02w1","Parents_Attitude_Data.YFAM2F03w1","Parents_Attitude_Data.YFAM2F04w1") # 역채점





# 또래관계
mean_frnd_feat = c("Friend_Data.YEDU2A01w1","Friend_Data.YEDU2A02w1","Friend_Data.YEDU2A03w1","Friend_Data.YEDU2A04w1",
              "Friend_Data.YEDU2A05w1","Friend_Data.YEDU2A06w1","Friend_Data.YEDU2A07w1")
# 교사관계
mean_tchr_feat = c("Teacher_Data.YEDU3A01w1","Teacher_Data.YEDU3A02w1","Teacher_Data.YEDU3A03w1","Teacher_Data.YEDU3A04w1",
              "Teacher_Data.YEDU3A05w1","Teacher_Data.YEDU3A06w1","Teacher_Data.YEDU3A07w1","Teacher_Data.YEDU3A08w1",
              "Teacher_Data.YEDU3A09w1","Teacher_Data.YEDU3A10w1","Teacher_Data.YEDU3A11w1","Teacher_Data.YEDU3A12w1",
              "Teacher_Data.YEDU3A13w1","Teacher_Data.YEDU3A14w1")
mean_hlplss_feat = c("Helplessness_Data.YINT2B01w1", "Helplessness_Data.YINT2B02w1", "Helplessness_Data.YINT2B03w1", "Helplessness_Data.YINT2B04w1",
                "Helplessness_Data.YINT2B05w1", "Helplessness_Data.YINT2B06w1", "Helplessness_Data.YINT2B07w1","Helplessness_Data.YINT2B08w1",
                "Helplessness_Data.YINT2B09w1","Helplessness_Data.YINT2B10w1", "Helplessness_Data.YINT2B11w1", "Helplessness_Data.YINT2B12w1",
                "Helplessness_Data.YINT2B13w1","Helplessness_Data.YINT2B14w1", "Helplessness_Data.YINT2B15w1", "Helplessness_Data.YINT2B16w1")

mean_grit_feat = c("Grit_Data.YPSY7A01w1", "Grit_Data.YPSY7A02w1", "Grit_Data.YPSY7A03w1","Grit_Data.YPSY7A04w1",
                  "Grit_Data.YPSY7A05w1","Grit_Data.YPSY7A06w1","Grit_Data.YPSY7A07w1","Grit_Data.YPSY7A08w1")
# 또래관계, 교사관계, 학업 무기력, 그릿, 자아존중감, 주의집중, 부모양육태도 각 영역의 평균 값을 각각 새로운 열로 만듦
result2 = add_mean_columns(result,list(mean_frnd_feat,mean_tchr_feat,mean_hlplss_feat,mean_grit_feat,mean_self_feat,mean_attention_feat,mean_attd_feat))
result2 %>% head()
```

```r
# 특정 값 대체 함수
replace_values <- function(x) {
  temp <- as.character(x)  # factor → character 변환
  temp[temp == "2"] <- "99"  # 임시 값으로 2 → 99
  temp[temp == "1"] <- "1"   # 1은 그대로 유지
  temp[temp == "3"] <- "2"   # 3 → 2
  temp[temp == "99"] <- "3"  # 임시 값 99 → 3
  return(as.factor(temp))   # 다시 factor로 변환
}


target_cols <- c("Profile_3")
selected_data[, (target_cols) := lapply(.SD, replace_values), .SDcols = target_cols]


# 결과 출력
head(selected_data,10)
```
